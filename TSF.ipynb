```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Data importation from the url 

```python
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(url)
data.columns = ['Hours', 'Scores'] 
print(data)
```

### Visualisation of Hours vs Scores

```python
X = data[['Hours']]
y = data['Scores']
plt.scatter(X,y)
plt.title('Visualisation of Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

### Split the Data into Training and Testing Sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Create and Train the Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

### Make Predictions

```python
y_pred = model.predict(X_test)
```

### Visualize the Results 

```python
plt.scatter(X, y, color='green')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

### Predict the score for 9.25 hours of study

```python
hours_studied = 9.25
predicted_score = model.predict([[hours_studied]])
print("Predicted Score for 9.25 hours of study: ", predicted_score[0])
```
